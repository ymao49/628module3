```python
import csv
import re
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python

with open('E:/stat628/module3/review_split.csv',newline = '', encoding='utf8')as f:
    data = csv.reader(f, delimiter=',')
    stars = []
    texts = []
    
    for row in data:
        star = row[3]
        text = row[7]
        stars.append(star)
        texts.append(text)

stars = stars[1:]    
texts = texts[1:]
```

```python
words = []
for txt in texts:
    x = re.findall("[a-z]+", txt)
    words.append(x)
```

```python
k = 0
high_review = []
low_review = []
for num in stars:
    if num == '4.0' or num == '5.0':
        high_review.append(words[k])
    if num == '1.0' or num == '2.0':
        low_review.append(words[k])
    k = k + 1
```

```python
import itertools

high_re = itertools.chain.from_iterable(high_review)
low_re = itertools.chain.from_iterable(low_review)
```

```python
from collections import Counter

high_re = Counter(high_re).most_common()
low_re = Counter(low_re).most_common()
```

```python
low = dict(low_re)
high = dict(high_re)
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").fit_words(low)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").fit_words(low)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
