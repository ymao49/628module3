```python
import json
import os
import pandas as pd
import numpy as np
os.chdir('/Users/pinecone/Downloads/Stat628')
```

```python
with open('tip.json', encoding='utf-8') as f:
    iter_f = iter(f)
    line = f.readline()
    tip = [list(eval(line).values())]
    for line in iter_f: 
        d = json.loads(line)
        tip.append(list(d.values()))
    f.close()
```

```python
col_names = list(eval(line).keys())
tip_df = pd.DataFrame(tip, columns=col_names)
```

```python
tip_df.head()
```

```python
with open('business.json', encoding='utf-8') as f:
    iter_f = iter(f)
    line = f.readline()
    business = [list(eval(line.replace("null", 'np.nan')).values())]
    for line in iter_f: 
        d = json.loads(line)
        business.append(list(d.values()))
    f.close()
```

```python
col_names = list(eval(line).keys())
business_df = pd.DataFrame(business, columns=col_names)
```

```python
business_df.head()
```

```python
categories = business_df.categories.tolist()
```

```python
while None in categories:
    categories.remove(None)
categories = ','.join(categories)
```

```python
categories = categories.split(',')
```

```python
categories = [elem.strip() for elem in categories]
```

```python
from collections import Counter
count_res = Counter(categories)
```

```python
count_res.most_common(100)
```
