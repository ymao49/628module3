```python
import json
import os
import pandas as pd
import numpy as np
```

## Load business.json and select rows that categories includes 'Steakhouses'.

```python
with open('business.json', encoding='utf-8') as f:
    iter_f = iter(f)
    business = []
    for line in iter_f: 
        d = json.loads(line)
        if d['categories'] and 'Steakhouses' in d['categories']:
            business.append(list(d.values()))
    f.close()
```

```python
col_names = list(eval(line).keys())
business_df = pd.DataFrame(business, columns=col_names)
```

```python
business_df.head()
```

```python
business_df.to_csv('business.csv', index=False)
```

## Save the business id to filter other json files.

```python
business_id = business_df.business_id.tolist()
```

## Load tip.json

```python
with open('tip.json', encoding='utf-8') as f:
    iter_f = iter(f)
    tip = []
    for line in iter_f: 
        d = json.loads(line)
        if d['business_id'] and d['business_id'] in business_id:
            tip.append(list(d.values()))
    f.close()
```

```python
col_names = list(eval(line).keys())
tip_df = pd.DataFrame(tip, columns=col_names)
```

```python
tip_df.to_csv('tip.csv', index=False)
```

```python
with open('review.json', encoding='utf-8') as f:
    iter_f = iter(f)
    review = []
    for line in iter_f: 
        d = json.loads(line)
        if d['business_id'] in business_id:
            review.append(list(d.values()))
    f.close()
```
